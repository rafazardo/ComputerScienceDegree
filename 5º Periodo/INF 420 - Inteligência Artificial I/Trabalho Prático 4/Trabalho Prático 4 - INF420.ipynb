```python
# Import de bibliotecas que serao usadas nos experimentos
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

from sklearn.naive_bayes import GaussianNB
from sklearn import tree
from sklearn import svm
from sklearn import neighbors
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

# Recursos de estilizacao
STYLE = "crest" # Style padrao para todos recursos de estilizacao
plt.style.use('bmh') # Style padrao para todos recursos do plt

# Componentes globais dos experimentos
melhoresAcuracias = [None, None, None, None, None, None, None] # Armazena a melhor acuracia obitida em cada experimento
metodosTestados = ["NB", "DT", "SVML", "SVMR", "k-NN", "RF", "GTB"] # Armazena as abreviações dos métodos testados
```

# Dataframe
Antes de continuar com a implementação, é necessário carregar o arquivo "koi_data.csv". Certifique-se de que o arquivo esteja localizado na mesma pasta deste notebook para que o dataframe possa ser carregado corretamente.

```python
# Coleta o dataframde do arquivo
dataframePrincipal = pd.read_csv('koi_data.csv')

# Separa os dataframes
features = dataframePrincipal.drop(['kepoi_name','koi_disposition'],axis=1)
targets = dataframePrincipal['koi_disposition']                           
```

# Treino com Dataframe
Antes de avançar com a implementação, é necessário criar as variáveis de treinamento e teste para as features e targets. Essas variáveis serão usadas posteriormente no processo de treinamento e avaliação do modelo. Certifique-se de preparar corretamente essas variáveis antes de prosseguir.

```python
# Realiza o treino e teste das features e targets
featuresTrain, featuresTest, targetsTrain, targetsTest = train_test_split(features, targets, test_size=0.5, random_state=0)
```

# Naive Bayes (NB)

```python
classificador = GaussianNB()

# Calcula o fit
targetsFit = classificador.fit(featuresTrain, targetsTrain)

# Calcula o predict
targetsPredict = targetsFit.predict(featuresTest)

# Coleta os dados da classificacao
resumoClassificacao = classification_report(targetsTest, targetsPredict, output_dict=True)
dfResumoClassificacao = pd.DataFrame(resumoClassificacao).transpose() # Armazena em um df do panda

# Calcula acuracia
acuracia = accuracy_score(targetsTest, targetsPredict)

# Calcula validacao cruzada
validacaoCruzada = cross_val_score(classificador, featuresTrain, targetsTrain, cv=5)

melhoresAcuracias[0] = acuracia # Adiciona a melhor acoracia obtido nos experimentos
```

## Gráficos de Resultados
Os testes realizados utilizando o método Naive Bayes nos forneceram uma valiosa oportunidade de avaliar o desempenho desse algoritmo de classificação. Por meio desses testes, foram gerados os seguintes gráficos, que nos ajudarão a interpretar os resultados obtidos.

### Classificação

```python
dfResumoClassificacaoEstilizada = dfResumoClassificacao.style.background_gradient(cmap=STYLE).format("{:.3f}") # Estiliza meu df

print("Resumo da Classificação:")
display(dfResumoClassificacaoEstilizada)
```

### Matriz de Confusão

```python
# Obtem a matriz de confusao
confusionMatrix = confusion_matrix(targetsTest, targetsPredict)

# Desenha a matriz de confusao
plt.figure(figsize=(3, 3))
sns.heatmap(confusionMatrix, annot=True, fmt=".1f", cmap=STYLE)
plt.title('Matriz de Confusão')
plt.xlabel('Rótulo Falso Verdadeiro')
plt.ylabel('Rótulo Verdadeiro')
plt.grid(False)
plt.show()

# Calcula os rotulos
totalRotulos = featuresTest.shape[0]
corretamenteRotulados = (targetsTest == targetsPredict).sum()
incorretamenteRotulados = (targetsTest != targetsPredict).sum()

print("Número de pontos corretamente rotulados de um total de", totalRotulos, "pontos:", corretamenteRotulados, "(", round((corretamenteRotulados / totalRotulos) * 100), "%)")
print("Número de pontos incorretamente rotulados de um total de", totalRotulos, "pontos:", incorretamenteRotulados, "(", round((incorretamenteRotulados / totalRotulos) * 100), "%) \n")
```

### Acurácia e Validação Cruzada K-fold

```python
dfResultadosAcuracia = pd.DataFrame({
    'Acurácia': [acuracia]
})

dfResultadosValidacaoCruzada = pd.DataFrame({
    'Validação Cruzada': validacaoCruzada  
})

dfResultadosAcuraciaEstilizada = dfResultadosAcuracia.style.background_gradient(cmap=STYLE).format("{:.3f}") # Estiliza meu df
dfResultadosValidacaoCruzadaEstilizada = dfResultadosValidacaoCruzada.style.background_gradient(cmap=STYLE).format("{:.3f}") # Estiliza meu df

print("Resumo da Acurácia e Validação Cruzada (k=5)")
display(dfResultadosAcuraciaEstilizada)
display(dfResultadosValidacaoCruzadaEstilizada)
```

Podemos observar que CONFIRMED teve uma alta taxa de recall (cerca de 97.5%) em comparação com a FALSE POSITIVE (cerca de 66.6%). Isso significa que o modelo tem uma melhor capacidade de identificar corretamente as instâncias CONFIRMED em relação FALSE POSITIVE. 

A acurácia geral do modelo foi de (cerca de 79%), o que indica um desempenho razoável na classificação das amostras. Além disso, a validação cruzada com k-fold (k=5) ressultou em uma média de acurácia de aproximadamente 0.812. # Decision Tree (DT)

```python
# Dados dos experimentos
tamanhosMaximosArvore = [50, 75, 100, 250, 300, 400, 500, "INFINITO"]

# Estrutura para armazenar resultados dos experimentos
targetsFits = []
targetsPredicts = []
dfResumosClassificacao = []
acuracias = []
validacoesCruzadas = []

for tamanhoMaximoArvore in tamanhosMaximosArvore:
    if (tamanhoMaximoArvore != "INFINITO"):
        classificador = tree.DecisionTreeClassifier(max_depth=tamanhoMaximoArvore)
    else:
        classificador = tree.DecisionTreeClassifier(max_depth=None) # Se for infinito, nao temos max_depth

    # Calcula o fit
    targetsFit = classificador.fit(featuresTrain, targetsTrain)
    targetsFits.append(targetsFit)
    
    # Calcula o predict
    targetsPredict = targetsFit.predict(featuresTest)
    targetsPredicts.append(targetsPredict)

    # Coleta os dados da classificacao
    resumoClassificacao = classification_report(targetsTest, targetsPredict, output_dict=True)
    dfResumoClassificacao = pd.DataFrame(resumoClassificacao).transpose() # Armazena em um df do panda
    dfResumosClassificacao.append(dfResumoClassificacao)
    
    # Calcula acuracia
    acuracia = accuracy_score(targetsTest, targetsPredict)
    acuracias.append(acuracia)
    
    # Calcula validacao cruzada
    validacaoCruzada = cross_val_score(classificador, featuresTrain, targetsTrain, cv=5)
    validacoesCruzadas.append(validacaoCruzada)
    
melhoresAcuracias[1] = acuracia # Adiciona a melhor acuracia obtido nos experimentos
```

# k-NN

```python
# Dados dos experimentos
numerosNeighbors = [50, 75, 100, 250, 300, 400, 450,500]

# Estrutura para armazenar resultados dos experimentos
targetsFits = []
targetsPredicts = []
dfResumosClassificacao = []
acuracias = []
validacoesCruzadas = []

for numeroNeighbors in numerosNeighbors:
    classificador = neighbors.KNeighborsClassifier(numeroNeighbors, weights='uniform')

    # Calcula o fit
    targetsFit = classificador.fit(featuresTrain, targetsTrain)
    targetsFits.append(targetsFit)
    
    # Calcula o predict
    targetsPredict = targetsFit.predict(featuresTest)
    targetsPredicts.append(targetsPredict)

    # Coleta os dados da classificacao
    resumoClassificacao = classification_report(targetsTest, targetsPredict, output_dict=True)
    dfResumoClassificacao = pd.DataFrame(resumoClassificacao).transpose() # Armazena em um df do panda
    dfResumosClassificacao.append(dfResumoClassificacao)
    
    # Calcula acuracia
    acuracia = accuracy_score(targetsTest, targetsPredict)
    acuracias.append(acuracia)
    
    # Calcula validacao cruzada
    validacaoCruzada = cross_val_score(classificador, featuresTrain, targetsTrain, cv=5)
    validacoesCruzadas.append(validacaoCruzada)
    
melhoresAcuracias[4] = acuracia # Adiciona a melhor acuracia obtido nos experimentos
```
