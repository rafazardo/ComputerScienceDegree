# Recursos Auxiliares
Antes de prosseguir com a implementação, é importante executar a célula a seguir para declarar as variáveis e recursos necessários que serão utilizados ao longo do código. Essa etapa é fundamental para garantir que todas as dependências estejam corretamente definidas e disponíveis para serem utilizadas no restante do código. Certifique-se de executar essa célula antes de continuar com a implementação.

```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

from sklearn.naive_bayes import GaussianNB
from sklearn import tree
from sklearn import svm
from sklearn import neighbors
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

STYLE = "crest" # Style padrao para todos recursos de estilizacao
plt.style.use('bmh') # Style padrao para todos recursos do plt

melhoresAcuracias = [] # Armazena a melhor acuracia obitida em cada experimento
metodosTestados = ["NB", "DT", "SVML", "SVMR", "k-NN", "RF", "GTB"]
```

# Dataframe
Antes de continuar com a implementação, é necessário carregar o arquivo "koi_data.csv". Certifique-se de que o arquivo esteja localizado na mesma pasta deste notebook para que o dataframe possa ser carregado corretamente.

```python
dataframePrincipal = pd.read_csv('koi_data.csv')

# Separa os dataframes
# dataframeFeatures = dataframePrincipal.drop(['kepoi_name','koi_disposition'],axis=1)
# dataframeTarget = dataframePrincipal['koi_disposition']

#features = dataframeFeatures # X
#targets = dataframeTarget # Y

X = dataframePrincipal.drop(['kepoi_name','koi_disposition'],axis=1)
y = dataframePrincipal['koi_disposition']                            
```

# Treino com Dataframe
Antes de avançar com a implementação, é necessário criar as variáveis de treinamento e teste para as features e targets. Essas variáveis serão usadas posteriormente no processo de treinamento e avaliação do modelo. Certifique-se de preparar corretamente essas variáveis antes de prosseguir.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
```

# Naive Bayes (NB)

```python
gnb = GaussianNB()

y_fit = gnb.fit(X_train, y_train)
y_pred = y_fit.predict(X_test)

resumoClassificacao = classification_report(y_test, y_pred, output_dict=True)
dfResumoClassificacao = pd.DataFrame(resumoClassificacao).transpose() # Armazena em um df do panda

acuracia = accuracy_score(y_test, y_pred)
validacaoCruzada = cross_val_score(gnb, X_train, y_train, cv=5)

melhoresAcuracias.append(acuracia)

# accuracy = gnb.score(X_test, y_test)
# print("Precisão:", accuracy)
```

## Gráficos de Resultados
Os testes realizados utilizando o método Naive Bayes nos forneceram uma valiosa oportunidade de avaliar o desempenho desse algoritmo de classificação. Por meio desses testes, foram gerados os seguintes gráficos, que nos ajudarão a interpretar os resultados obtidos.

### Classificação

```python
dfResumoClassificacaoEstilizada = dfResumoClassificacao.style.background_gradient(cmap=STYLE).format("{:.3f}") # Estiliza meu df

print("Resumo da Classificação:")
display(dfResumoClassificacaoEstilizada)
```

### Matriz de Confusão

```python
# Obtem a matriz de confusão
confusionMatrix = confusion_matrix(y_test, y_pred)

# Desenha a matriz de confusão
plt.figure(figsize=(3, 3))
sns.heatmap(confusionMatrix, annot=True, fmt=".1f", cmap=STYLE)
plt.title('Matriz de Confusão')
plt.xlabel('Rótulo Falso Verdadeiro')
plt.ylabel('Rótulo Verdadeiro')
plt.show()

print("Número de pontos incorretamente rotulados de um total de ",X_test.shape[0]," pontos: ",(y_test != y_pred).sum())
```

### Acurácia e Validação Cruzada K-fold

```python
dfResultadosAcuracia = pd.DataFrame({
    'Acurácia': [acuracia]
})

dfResultadosValidacaoCruzada = pd.DataFrame({
'Validação Cruzada': validacaoCruzada  
})

dfResultadosAcuraciaEstilizada = dfResultadosAcuracia.style.background_gradient(cmap=STYLE).format("{:.3f}") # Estiliza meu df
dfResultadosValidacaoCruzadaEstilizada = dfResultadosValidacaoCruzada.style.background_gradient(cmap=STYLE).format("{:.3f}") # Estiliza meu df

print("Resumo da Acurácia e Validação Cruzada (k=5)")
display(dfResultadosAcuraciaEstilizada)
display(dfResultadosValidacaoCruzadaEstilizada)
```

## Inferência Sobre os Resultados


# Decision Tree (DT)

```python
tamanhosMaximosArvore = [10, 20, 30, 40, 50, 60, 70, "INFINITO"]
y_fits = []
y_preds = []
dfResumosClassificacao = []
acuracias = []
validacoesCruzadas = []

for tamanhoMaximoArvore in tamanhosMaximosArvore:

    if (tamanhoMaximoArvore != "INFINITO"):
        clf = tree.DecisionTreeClassifier(max_depth=tamanhoMaximoArvore)
    else:
        clf = tree.DecisionTreeClassifier(max_depth=None)

    y_fit = clf.fit(X_train, y_train)
    y_fits.append(y_fit)
    
    y_pred = y_fit.predict(X_test)
    y_preds.append(y_pred)

    resumoClassificacao = classification_report(y_test, y_pred, output_dict=True)
    dfResumoClassificacao = pd.DataFrame(resumoClassificacao).transpose() # Armazena em um df do panda
    dfResumosClassificacao.append(dfResumoClassificacao)
    
    acuracia = accuracy_score(y_test, y_pred)
    acuracias.append(acuracia)
    
    validacaoCruzada = cross_val_score(clf, X_train, y_train, cv=5)
    validacoesCruzadas.append(validacaoCruzada)
    
melhoresAcuracias.append(max(acuracias))
```

## Gráficos de Resultados

Os testes realizados utilizando o método Decision Tree nos forneceram uma valiosa oportunidade de avaliar o desempenho desse algoritmo de classificação. Por meio desses testes, foram gerados os seguintes gráficos, que nos ajudarão a interpretar os resultados obtidos.

### Classificação

```python
for i in range(len(dfResumosClassificacao)):
    dfResumoClassificacaoEstilizada = dfResumosClassificacao[i].style.background_gradient(cmap=STYLE).format("{:.3f}") # Estiliza meu df

    print("Resumo da classificação para arvore de tamanho maximo", tamanhosMaximosArvore[i])
    display(dfResumoClassificacaoEstilizada)
```

### Matriz de Confusão

```python
for i in range(len(dfResumosClassificacao)):
    print("Matriz de confusão para arvore de tamanho maximo", tamanhosMaximosArvore[i])
    
    # Obtem a matriz de confusão
    confusionMatrix = confusion_matrix(y_test, y_preds[i])

    # Desenha a matriz de confusão
    plt.figure(figsize=(3, 3))
    sns.heatmap(confusionMatrix, annot=True, fmt=".1f", cmap=STYLE)
    plt.title('Matriz de Confusão')
    plt.xlabel('Rótulo Falso Verdadeiro')
    plt.ylabel('Rótulo Verdadeiro')
    plt.show()

    print("Número de pontos incorretamente rotulados de um total de ",X_test.shape[0]," pontos: ",(y_test != y_preds[i]).sum(),"\n")
```

### Acurácia e Validação Cruzada K-fold

```python
dfResultadosAcuracia = pd.DataFrame({
'Acurácia': acuracias
}, index = tamanhosMaximosArvore)

mediasValidacoesCruzadas = [np.mean(k5Values) for k5Values in validacoesCruzadas]

dfResultadosValidacaoCruzada = pd.DataFrame({
'Validação Cruzada (média)': mediasValidacoesCruzadas
}, index = tamanhosMaximosArvore)


dfResultadosAcuraciaEstilizada = dfResultadosAcuracia.style.background_gradient(cmap=STYLE).format("{:.3f}") # Estiliza meu df
dfResultadosValidacaoCruzadaEstilizada = dfResultadosValidacaoCruzada.style.background_gradient(cmap=STYLE).format("{:.3f}") # Estiliza meu df

display(dfResultadosAcuraciaEstilizada)
display(dfResultadosValidacaoCruzadaEstilizada)
```

### Comparação da Acurácia e Validação Cruzada K-fold

```python
plt.plot(tamanhosMaximosArvore, acuracias, marker='s')
plt.xlabel('Número de Árvores')
plt.ylabel('Acurácia')
plt.title('Variação da Acurácia em relação ao Número de Árvores')
plt.grid(False)
plt.show()

plt.plot(tamanhosMaximosArvore, validacoesCruzadas, marker='s')
plt.xlabel('Número de Árvores')
plt.ylabel('Validação Cruzada')
plt.title('Variação da Validação Cruzada em relação ao Número de Árvores')
plt.grid(False)
plt.show()
```

## Inferência Sobre os Resultados

# SVM Linear (SVML)

```python
clf = svm.SVC(kernel='linear')

y_fit = clf.fit(X_train, y_train)
y_pred = y_fit.predict(X_test)

accuracy = clf.score(X_test, y_test)

print("LINEAR KERNEL")
print("Precisão:", accuracy)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
```

# SVM RBF (SVMR)

```python
clf = svm.SVC(kernel='rbf')

y_fit = clf.fit(X_train, y_train)
y_pred = y_fit.predict(X_test)

accuracy = clf.score(X_test, y_test)

print("RBF KERNEL")
print("Precisão:", accuracy)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
```

# k-NN

```python
numNeighbors = 10

clf = neighbors.KNeighborsClassifier(numNeighbors, weights='uniform')

y_fit = clf.fit(X_train, y_train)
y_pred = y_fit.predict(X_test)

accuracy = clf.score(X_test, y_test)

print("Precisão:", accuracy)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
```

# Random Forest (RF)

```python
numArvores = 25

clf = RandomForestClassifier(n_estimators=numArvores)

y_fit = clf.fit(X_train, y_train)
y_pred = y_fit.predict(X_test)

accuracy = clf.score(X_test, y_test)

print("Precisão:", accuracy)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
```

# Gradient Tree Boosting (GTB)

```python
numIteracoes = 100

clf = GradientBoostingClassifier(n_estimators=numIteracoes)

y_fit = clf.fit(X_train, y_train)
y_pred = y_fit.predict(X_test)

accuracy = clf.score(X_test, y_test)

print("Precisão:", accuracy)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
```
